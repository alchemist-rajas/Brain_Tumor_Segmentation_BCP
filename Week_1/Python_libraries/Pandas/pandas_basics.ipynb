```python
# Assignment 1
'''
Import videogamesales.csv and perform the following actions on it
Add a column of 'global_sales' showing the total sales of all the different regions to the data frame and sort (highest first) and print the DataFrame according to it.
Display a plot of the total number of copies sold of each genre globally.
Filter out only the games containing 'Grand Theft Auto' in their name and display the following as a DataFrame:
a. their name
b. the platform they were released on
c. the year they were released in
d. the sum of sales in only Europe and Japan
Display a pie chart of the total sales of all Grand Theft Auto games combined in North America, Europe, Japan, and other sales.
'''
```

```python
# TODO
def pandas_assignment_1():
    # Assuming the file name is 'videogames.csv' based on the prompt's context.
    file_name = 'videogamesales.csv'
    
    try:
        df = pd.read_csv(file_name)
    except FileNotFoundError:
        print(f"Error: The file '{file_name}' was not found. Please ensure '{file_name}' or 'videogamesales.csv' is uploaded.")
        return None

    # Define sales columns and convert them to numeric
    sales_cols = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']
    for col in sales_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        
    # 1. Add 'Global_Sales' column
    df['Global_Sales'] = df[sales_cols].sum(axis=1)

    # 2. Sort (highest first) and print the DataFrame
    df_sorted = df.sort_values(by='Global_Sales', ascending=False)
    print("\n--- 1. DataFrame sorted by Global Sales (Highest First) ---")
    print(df_sorted.head())
    df_sorted.to_csv('videogames_sorted.csv', index=False)
    
    # 3. Display a plot of the total number of copies sold of each genre globally
    genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
    
    plt.figure(figsize=(12, 6))
    genre_sales.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Total Global Copies Sold by Game Genre', fontsize=16)
    plt.xlabel('Genre', fontsize=12)
    plt.ylabel('Total Global Sales (in Millions)', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.savefig('global_sales_by_genre.png')
    plt.close()
    print("\n--- 2. Plot of Total Global Sales by Genre Saved as 'global_sales_by_genre.png' ---")


    # 4. Filter out only the games containing 'Grand Theft Auto' in their name
    gta_df = df[df['Name'].str.contains('Grand Theft Auto', case=False, na=False)].copy()
    
    # Calculate the sum of sales in only Europe and Japan
    gta_df['EU_JP_Sales'] = gta_df['EU_Sales'] + gta_df['JP_Sales']
    
    # 5. Display the filtered DataFrame
    gta_output_df = gta_df[['Name', 'Platform', 'Year', 'EU_JP_Sales']]
    
    print("\n--- 3. Filtered 'Grand Theft Auto' Games Data ---")
    print(gta_output_df)
    gta_output_df.to_csv('gta_filtered_data.csv', index=False)
    
    # 6. Display a pie chart of the total sales of all Grand Theft Auto games combined by region
    gta_regional_sales = gta_df[sales_cols].sum()
    
    pie_labels = {
        'NA_Sales': 'North America',
        'EU_Sales': 'Europe',
        'JP_Sales': 'Japan',
        'Other_Sales': 'Other'
    }
    
    sales_data = gta_regional_sales.values
    sales_labels = [pie_labels[col] for col in gta_regional_sales.index]
    
    plt.figure(figsize=(8, 8))
    plt.pie(sales_data, labels=sales_labels, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors, wedgeprops={'edgecolor': 'black'})
    plt.title('Total Grand Theft Auto Sales by Region', fontsize=14)
    plt.tight_layout()
    plt.savefig('gta_regional_sales_pie.png')
    plt.close()
    print("\n--- 4. Pie Chart of GTA Regional Sales Saved as 'gta_regional_sales_pie.png' ---")
    
    return None

pandas_assignment_1()
```
